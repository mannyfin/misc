```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def coin_toss(n):
    # heads is 0, tails is 1
    sample = np.random.randint(0,2,size=n)
    num_heads = sum((sample<1)*1)
    num_tails = sum((sample==1)*1)
    return num_heads, num_tails

def coin_minus_half(n):
    #     find proportion of heads - 1/2
    num_heads, num_tails = coin_toss(n)
    return (num_heads/(n) - 0.5), (num_heads-0.5*n)

trials = 10000
a_list=[]
b_list=[]
for flips in range(1,trials):
    a, b = coin_minus_half(flips)
    a_list.append(a)
    b_list.append(b)
```

```python
plt.figure(1)
plt.title('proportion of heads minus 1/2')
plt.plot(a_list, color='k')
plt.figure(2)
plt.title('number of heads minus half the number of tosses')
plt.plot(b_list, color='r')
plt.legend
plt.show()
```

```python
a_list = np.array(a_list)
# 10 percent variance around 0.5 probability
print(np.where(a_list > 0.1))

print(np.where(a_list < -0.1))

print(len(np.where(a_list > 0.1)[0]) + len(np.where(a_list < -0.1)[0]))
# for a 95 percent chance proportion is between 0.4 and 0.6, the number of flips must be...around 500 but your mileage will vary
print((len(np.where(a_list > 0.1)[0]) + len(np.where(a_list < -0.1)[0])) / 0.05)
```
